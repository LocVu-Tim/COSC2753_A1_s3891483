## 2. Data Distribution

### 2.1 Overview
##### We will observe the data distribution by creating histograms for each column in our datasets. The test and train dataset will be overlaid for better analyzing and comparing. 

#### 2.1.1 Train Dataset

#### Train Histogram Plot:

```python
plt.figure(figsize=(25,25))
for i, col in enumerate(df_train.columns):
    plt.subplot(5,6,i+1)
    plt.hist(df_train[col], alpha=0.3, color='b', density=True)
    plt.title(col)
    plt.xticks(rotation='vertical')
```

```python

fig, axes = plt.subplots(5, 5, figsize=(10, 8))

for i, ax in enumerate(axes.ravel()):
    if i < len(df_train.columns):
        ax.hist(df_train.iloc[:, i], bins=15)
        ax.set_xlabel(df_train.columns[i])
        ax.set_ylabel('Frequency')
        ax.axvline(df_train.iloc[:, i].mean(),alpha=0.3, color='r', linestyle='dashed', linewidth=1)
        ax.axvline(df_train.iloc[:, i].median(),alpha=0.3, color='g', linestyle='dashed', linewidth=1)
    else:
        ax.axis('off')

plt.tight_layout() 
plt.show()
```

#### Train Correlation Matrix

```python
f, ax = plt.subplots(figsize=(11, 9))
corr = df_train.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=90,
    horizontalalignment='right'
);
```

#### 2.1.2 Test Dataset

#### Test Histogram

```python
plt.figure(figsize=(25,25))
for i, col in enumerate(df_test.columns):
    plt.subplot(5,6,i+1)
    plt.hist(df_test[col], alpha=0.3, color='b', density=True)
    plt.title(col)
    plt.xticks(rotation='vertical')
```

```python
fig, axes = plt.subplots(5, 5, figsize=(10, 8))

for i, ax in enumerate(axes.ravel()):
    if i < len(df_test.columns):
        ax.hist(df_test.iloc[:, i], bins=15)
        ax.set_xlabel(df_test.columns[i])
        ax.set_ylabel('Frequency')
        ax.axvline(df_test.iloc[:, i].mean(), color='r', linestyle='dashed', linewidth=1)
        ax.axvline(df_test.iloc[:, i].median(), color='g', linestyle='dashed', linewidth=1)
    else:
        ax.axis('off')

plt.tight_layout() 
plt.show()
```

#### Test Correlation Matrix

```python
f, ax = plt.subplots(figsize=(11, 9))
corr = df_train.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=90,
    horizontalalignment='right'
);
```

#### Comparision between train and test dataset

```python
num_features = len(df_train.columns)
num_cols = min(num_features, 3)
num_rows = (num_features + num_cols - 1) // num_cols

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5*num_rows), constrained_layout=True)
axes = axes.flatten()

for i, column in enumerate(df_train.columns):
    ax = axes[i]
    sns.histplot(data=df_train, x=column, ax=ax, color='blue', alpha=0.5, label='Train', kde=True)
    sns.histplot(data=df_test, x=column, ax=ax, color='orange', alpha=0.5, label='Test', kde=True)
    ax.set_title(f"Histogram of {column}")
    ax.set_ylabel("Frequency")
    ax.set_xlabel(column)
    ax.legend()

for i in range(num_features, len(axes)):
    axes[i].remove()

plt.tight_layout()
plt.show()
```
